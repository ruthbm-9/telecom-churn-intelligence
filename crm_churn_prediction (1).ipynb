#**1. Market research and hypothesis**

The Latin American telecommunications industry faces significant financial challenges associated with customer churn, making retention a strategic priority. ***Annual churn rates are estimated to range between 20% and 40%***, ***while retaining existing customers is considerably more cost-effective than acquiring new ones, with retention costs estimated to be 5 to 10 times lower than acquisition costs*** (Abarca Sánchez et al., 2022). Furthermore, customer churn prediction has been identified as a critical capability because predicting and preventing customer attrition is estimated to be up to 16 times less costly than acquiring new customers to replace those who have left. ***Moreover, studies suggest that reducing the churn rate by just 5% can increase profitability by between 25% and 85%*** (Abarca Sánchez et al., 2022). The importance of churn management has intensified due to market saturation, increased customer mobility, and strong competitive pressure within the telecommunications sector, making churn prediction a key component of strategic planning and operational decision-making (Abarca Sánchez et al., 2022; Saha et al., 2023). In addition, op***erational inefficiencies continue to affect industry performance***, as evidenced by revenue declines reported in regional telecommunications markets and financial penalties associated with service-level failures (Correa, 2019; Mocoroa, 2025). Collectively, t***hese factors highlight customer churn as one of the most critical business challenges facing telecommunications providers today.***

## 1.2 Literature Review: Drivers of Customer Churn




For this reason, we conducted research on the main issues within the Latin American context, studying the primary reasons why customers switch operators:

***1.Customer Experience and Satisfaction***

Overall customer experience is a critical determinant of retention, as users tend to remain with providers that deliver superior perceived value (Correa, 2019). Key aspects include service quality, such as the empathy, professionalism, and guidance provided by customer service representatives (Abarca Sánchez et al., 2022). A behavioral inertia effect is also observed, where longer-tenured customers show higher resistance to switching providers, even when alternatives exist in the market (Correa, 2019).

2.***Technical Quality and Service Reliability***

Technical performance is a fundamental driver of customer retention. Efficient resolution of service failures and minimization of downtime significantly improve customer loyalty (Arteaga Ventura et al., 2024). Network coverage is also a critical component, as it determines the ability to maintain successful communications (Correa, 2019). Additionally, connection speed and service reliability are essential for maintaining continuity, particularly for high-demand users and corporate segments (Arteaga Ventura et al., 2024).

3. ***Market Incentives and Pricing Factors***

Perceived price fairness is a key retention factor, where customers are more likely to remain with operators whose pricing aligns with their economic expectations (Abarca Sánchez et al., 2022). Network effects also play an important role, as users tend to prefer providers with larger subscriber bases due to increased communication opportunities (Correa, 2019).

4. ***Demographic and Behavioral Factors***

Customer demographics, including age, gender, socioeconomic status, and spending levels, influence loyalty and service selection (Correa, 2019). Additionally, usage behavior and plan type (e.g., prepaid vs. postpaid) are strong determinants of customer permanence, as they reflect both engagement intensity and financial commitment (Correa, 2019).




##1.3 General hypothesis


In the Latin American telecommunications market, ***customer churn is not a purely technical phenomenon caused by signal failures. While deficiencies in coverage and network instability are the most visible triggers of dissatisfaction, the final decision to switch operators is conditioned by an ecosystem of surrounding latent factors***. These include the wear and tear of the relationship's lifecycle (tenure and hardware obsolescence), the perceived value relative to monthly spending (financial commitment), and the sociodemographic characteristics of the user's environment. Therefore, to effectively understand and predict churn, ***it is essential to analyze the customer experience from a holistic and multidimensional perspective***, as these peripheral factors can have an equal or even more definitive impact than the technical quality of the network.

### 3.3.1 Hypothesis 1

Technical issues are widely cited as a leading cause of churn. Yet, we must determine if this holds true for our base: are customers leaving primarily due to technical failures, or is there a 'non-technical' churn segment leaving for other underlying reasons?

***Do churned customers show higher technical problems compared with customers who stayed?***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot: technical issues vs churn

plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x="churn",
    y="drop_blk_Mean",
    hue="churn", # Assign x variable to hue
    palette={0: 'green', 1: 'red'},
    legend=False # Set legend to False to avoid duplicate legends
)

plt.xlabel("Customer churn (0 = Stayed, 1 = Churned)")
plt.ylabel("Average dropped/blocked calls")
plt.title("Technical Issues Distribution by Customer Churn")
# Customize the y-axis limits here (e.g., to focus on a specific range)
plt.ylim(-20, 20) # You can change these values as needed
plt.show()
```

Technical service issues appear to affect a subset of churned customers, especially those with extreme levels of failures. However, since the median technical issue level of churned customers is not higher than retained customers, technical degradation alone does not explain the majority of customer abandonment. This suggests the existence of silent churn driven by other factors.

```python

df["technical_group"] = pd.qcut(
    df["technical_issue_score"],
    q=2,
    labels=["Low technical issues", "High technical issues"]
)


churn_by_technical = (
    df.groupby("technical_group", observed=False)["churn"] # Added observed=False to address FutureWarning
    .mean()
    * 100
)

print(churn_by_technical)

# Plotting the bar chart with custom colors
plt.figure(figsize=(8, 5))

# Define colors for the technical groups as requested
colors = {
    "Low technical issues": "green",
    "High technical issues": "orange"
}

# Map the colors to the categories in churn_by_technical index
bar_colors = [colors[group] for group in churn_by_technical.index]

churn_by_technical.plot(
    kind='bar',
    color=bar_colors
)

plt.title('Churn Rate by Technical Issue Group')
plt.xlabel('Technical Issue Group')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=0) # Changed rotation to 0 for horizontal labels
plt.tight_layout()
plt.show()
```

The technical issue segmentation shows that churn is not strongly associated with service quality problems. Customers with low levels of technical issues present the highest churn rate, while customers with medium and high technical issues show slightly lower churn rates. This indicates that technical degradation alone does not explain customer abandonment and suggests the presence of silent churn, where customers leave due to other factors beyond network performance.



> ***Hypothesis 1 conclusion:***
Technical degradation is a visible but limited churn driver. A small group of customers may leave due to service quality issues; however, most churn appears to occur without strong evidence of technical problems, suggesting the presence of silent churn.

While technical issues and customer service interactions do not clearly differentiate churned customers, ***behavioral indicators such as declining usage and reduced customer activity show stronger associations with abandonment***.
```python
import pandas as pd
import matplotlib.pyplot as plt



exclude = [
    "churn",
    "Customer_ID"
]


numeric_cols = [
    col for col in df.select_dtypes(include="number").columns
    if col not in exclude
]




impact = []

for col in numeric_cols:

    churn_mean = df[df["churn"]==1][col].mean()
    stayed_mean = df[df["churn"]==0][col].mean()

    diff = churn_mean - stayed_mean

    impact.append(
        [col, diff]
    )


impact_df = pd.DataFrame(
    impact,
    columns=["Variable","Mean_difference"]
)




impact_df["abs_difference"] = (
    impact_df["Mean_difference"].abs()
)


top_variables = (
    impact_df
    .sort_values(
        "abs_difference",
        ascending=False
    )
    .head(20)
)




plt.figure(figsize=(10,8))

plt.barh(
    top_variables["Variable"],
    top_variables["Mean_difference"]
)

plt.xlabel(
    "Difference in mean (Churned - Stayed)"
)

plt.title(
    "Variables with the Largest Separation Between Churned and Stayed Customers"
)

plt.gca().invert_yaxis()

plt.show()
```

##8.1 Threshold sensitivity

When evaluating different threshold values, a trade-off was observed between Recall and Precision. A higher Recall reduces false negatives, allowing the model to identify a larger proportion of customers who are truly at risk of churning, but it also increases false positives and therefore the number of retention campaigns. Conversely, a higher Precision reduces unnecessary campaigns, but increases the risk of failing to identify customers who eventually leave the company.

For this reason, t***hreshold selection will not be based solely on classification metrics. Instead, a sensitivity analysis will be conducted to evaluate how different threshold values translate into economic outcomes***, comparing the cost of contacting customers who would not have churned against the cost of losing customers who actually churn.

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

import pandas as pd
import numpy as np

# Probabilidades en validation
y_prob_val = final_cat.predict_proba(X_val_int)[:, 1]

results = []

for t in np.arange(0.20, 0.61, 0.01):

    y_pred = (y_prob_val >= t).astype(int)

    tn, fp, fn, tp = confusion_matrix(
        y_val_int,
        y_pred
    ).ravel()

    results.append({
        "Threshold": round(t, 2),
        "Accuracy": accuracy_score(y_val_int, y_pred),
        "Precision": precision_score(y_val_int, y_pred),
        "Recall": recall_score(y_val_int, y_pred),
        "F1": f1_score(y_val_int, y_pred),
        "False_Positive": fp,
        "False_Negative": fn,
        "True_Positive": tp,
        "True_Negative": tn
    })

threshold_results = pd.DataFrame(results)

threshold_results.sort_values(
    "F1",
    ascending=False
).head(20)
```

**Business assumptions for economic evaluation**

* ***Retention campaign cost***: Based on Silva Vilchez (2024), the estimated customer retention cost in the mobile telecommunications sector was assumed ***as USD 19.78 per user***, representing the cost of customer intervention campaigns.
* ***Retention success rate***: A ***17% retention rate*** was assumed based on Silva Vilchez (2024), meaning that only 20% of customers correctly identified as potential churners are expected to remain active after the intervention.
* **Customer value estimation:** Customer value was estimated u***sing totmrc_Mean*** as monthly revenue and a ***conservative 6-month*** revenue horizon:
      
$$ Customer\ Value = Monthly\ Revenue \times 6
$$

* ***Economic break-even assumption***: The model creates v***alue when the revenue protected from correctly identified and retained customers exceeds the costs generated by model errors***.
  
    
$$
Economic\ Value = (TP \times Retention\ Rate \times Customer\ Value) - (FP\ Cost + FN\ Cost)
$$

where:

- TP = True Positives
- FP = False Positives
- FN = False Negatives

* **Error cost consideration:**

False Negatives (FN): represent customers who churn but are not detected, causing potential revenue loss.

False Positives (FP): represent unnecessary retention actions. Although these customers would not have churned, they generate additional operational costs due to customer service time and campaign resources.

```python
import pandas as pd
import numpy as np

from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

# -----------------------
# BUSINESS ASSUMPTIONS
# -----------------------

cost_fp = 19.78  # Cost Marketing campaign per user

monthly_revenue = 44.99
months_retained = 6

customer_value = monthly_revenue * months_retained

# Fixed campaign success rate
retention_rate = 0.17  # 17% of identified churners are retained

print("Cost FP:", cost_fp)
print("Customer Value:", customer_value)
print("Retention Rate:", retention_rate)


# -----------------------
# THRESHOLD ANALYSIS
# -----------------------

thresholds = np.arange(0.10, 0.91, 0.05)

# Probabilidades en validation
y_prob_val = final_cat.predict_proba(X_val_int)[:, 1]

results = []

for th in thresholds:

    y_pred = (y_prob_val >= th).astype(int)

    tn, fp, fn, tp = confusion_matrix(
        y_val_int,
        y_pred
    ).ravel()

    precision = precision_score(
        y_val_int,
        y_pred,
        zero_division=0
    )

    recall = recall_score(
        y_val_int,
        y_pred,
        zero_division=0
    )

    f1 = f1_score(
        y_val_int,
        y_pred,
        zero_division=0
    )

    # -----------------------
    # BUSINESS CALCULATIONS
    # -----------------------

    # Cost of unnecessary retention actions
    fp_cost = fp * cost_fp


    # Revenue loss from missed churners
    fn_cost = fn * customer_value


    # Total error cost
    total_error_cost = fp_cost + fn_cost


    # Customers actually retained
    retained_customers = tp * retention_rate


    # Value generated from retained customers
    retained_customer_value = (
        retained_customers * customer_value
    )


    # Final economic value
    economic_value = (
        retained_customer_value
        - total_error_cost
    )


    results.append([
        th,
        precision,
        recall,
        f1,
        tp,
        fp,
        fn,
        retained_customers,
        retained_customer_value,
        fp_cost,
        fn_cost,
        total_error_cost,
        economic_value
    ])


results_df = pd.DataFrame(
    results,
    columns=[
        "Threshold",
        "Precision",
        "Recall",
        "F1",
        "True_Positives",
        "False_Positives",
        "False_Negatives",
        "Retained_Customers",
        "Retained_Customer_Value",
        "FP_Cost",
        "FN_Cost",
        "Total_Error_Cost",
        "Economic_Value"
    ]
)


results_df = results_df.sort_values(
    "Economic_Value",
    ascending=False
)


results_df.head(20)
```

The retention decision tree transforms the churn model output into a business action strategy. # Risk Segmentation Layer
Customers are grouped according to the probability assigned by the model.

The probability represents the estimated likelihood of churn.

```python
def risk_category(prob):

    if prob >= 0.80:
        return "High Risk"

    elif prob >= 0.50:
        return "Medium Risk"

    else:
        return "Low Risk"



retention_df["risk_level"] = (
    retention_df["churn_probability"]
    .apply(risk_category)
)



retention_df["risk_level"].value_counts()
```

#Customer value segmentation
Customer value is divided using the distribution of totmrc_Mean:

Low Value: below P50
Medium Value: P50–P75
High Value: above P75

```python
# Add customer value

retention_df["customer_value"] = (
    retention_df["totmrc_Mean"]
)



# Percentiles

p50 = retention_df["customer_value"].quantile(0.50)

p75 = retention_df["customer_value"].quantile(0.75)



def value_category(value):

    if value < p50:
        return "Low Value"

    elif value < p75:
        return "Medium Value"

    else:
        return "High Value"



retention_df["value_level"] = (
    retention_df["customer_value"]
    .apply(value_category)
)



retention_df["value_level"].value_counts()
```

#Retention priority

```python
def assign_priority(row):

    if (
        row["risk_level"] == "High Risk"
        and row["value_level"] == "High Value"
    ):
        return "Priority 1"


    elif (
        row["risk_level"] == "High Risk"
        and row["value_level"] == "Medium Value"
    ):
        return "Priority 2"


    elif (
        row["risk_level"] == "Medium Risk"
        and row["value_level"] == "High Value"
    ):
        return "Priority 3"


    elif (
        row["risk_level"] == "High Risk"
        and row["value_level"] == "Low Value"
    ):
        return "Priority 4"


    else:
        return "Priority 5"



retention_df["retention_priority"] = (
    retention_df
    .apply(assign_priority, axis=1)
)
```

To determine the main drivers contributing to churn prediction, a comparison between CatBoost feature importance and SHAP importance was performed.

Although both approaches measure feature relevance, their interpretation is different. C***atBoost feature importance reflects how frequently or effectively a variable is used by the model during the tree-building process. In contrast, SHAP importance measures the actual contribution of each feature*** to the model output by quantifying how much each variable influences the final prediction.

Th***e comparison showed that some variables presented differences between both rankings. A feature may appear highly important according to CatBoost because it contributes to the creation of decision rules within the trees; however, its actual impact on the final predictions may be lower when evaluated through SHAP values***.

***For this reason, SHAP importance was selected as the main criterion for identifying the variables included in the Customer Retention Intelligence Layer***.
